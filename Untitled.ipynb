## Importing packages

```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns #for creating heat maps
import numpy as np
```

## Reading the data

Now we will read 6 csv files that contain the data from April 2014 to September 2014. After reading the data, we will combine all of them into a single dataframe called 'data'.

```python
apr_data = pd.read_csv('uber-raw-data-apr14.csv')
may_data = pd.read_csv('uber-raw-data-may14.csv')
jun_data = pd.read_csv('uber-raw-data-jun14.csv')
jul_data = pd.read_csv('uber-raw-data-jul14.csv')
aug_data = pd.read_csv('uber-raw-data-aug14.csv')
sep_data = pd.read_csv('uber-raw-data-sep14.csv')
```

```python
data = pd.concat([apr_data, may_data, jun_data, jul_data, aug_data, sep_data])
```

```python
data.head()
```

Now we will create columns like 'Time', 'Date', 'Hour', 'Month', 'Day', 'Year'.

```python
data['Time'] = data['Date/Time'].str.slice(-8,-6).astype(int)
```

```python
data.Time.value_counts(sort = False)
```

```python
data['Date'] = data['Date/Time'].str.slice(0,9)
```

```python
data.Date = pd.to_datetime(data.Date)
data['Date/Time'] = pd.to_datetime(data['Date/Time'])
```

```python
data['Month'] = data.Date.dt.strftime('%b')
data['Day'] = data.Date.dt.strftime('%d')
data['Year'] = data.Date.dt.strftime('%y')
```

```python
data['dayofweek'] = data['Date/Time'].dt.weekday_name
```

```python
data['WEEKDAY'] = data['dayofweek'].apply(lambda x : 'Weekend' if (x == 'Sunday' or x == 'Saturday') else 'Weekday')
```

```python
data.head()
```

## Plotting the trips by the hours in a day


```python
df = data.groupby(['Time', 'Month'])['Time'].count().unstack('Month')
```

```python
df.head()
```

```python
%matplotlib inline
```

```python
sns.set()
df.plot(kind = 'bar', stacked = True, title = 'Trips by Hour and Mounth')
plt.ylabel('Total')
```
